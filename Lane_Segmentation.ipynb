```
from __future__ import print_function
import numpy as np 
import os
import glob
import skimage.io as io
import skimage.transform as trans
from keras.preprocessing.image import ImageDataGenerator
from keras.models import *
from keras.layers import *
from keras.optimizers import *
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras.metrics import MeanIoU as mio
import keras.preprocessing
from keras import backend as K
import cv2
from PIL import Image as im 
import matplotlib.pyplot as plt
```

```
batch_size = 8
train_path = '/content/train'
validate_path = '/content/validate'
aug_folder_img = '/content/train/aug/images'
aug_folder_mask = '/content/train/aug/onlylabels'
target_size = (240,320)
seed = 2
a = cv2.imread(os.path.join(train_path,'onlylabels/1573372397.46.png'))
gray = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)/255
for pixel in gray:
  pixel[pixel > 0] = 1
  pixel[pixel == 0] = 0
plt.imshow(gray)
print(gray.shape)
```

```
#Augmented-Training-Data
def augTrain():
  image_datagen = ImageDataGenerator()
  image_generator = image_datagen.flow_from_directory(train_path,
                                                      classes = ['images'],
                                                      class_mode = None,
                                                      color_mode = 'rgb',
                                                      target_size = target_size,
                                                      batch_size = batch_size,
                                                      seed = seed)
  mask_generator = image_datagen.flow_from_directory(train_path,
                                                    classes = ['onlylabels'],
                                                    class_mode = None,
                                                    color_mode = 'grayscale',
                                                    target_size = target_size,
                                                    batch_size = batch_size,
                                                    seed = seed)
  train_generator = zip(image_generator, mask_generator)
  i = 0
  for (image,mask) in train_generator:
    image = image / 255
    mask[mask > 0] = 1
    mask[mask == 0] = 0
    mask = np.squeeze(mask,axis = 3).astype(np.int32)
    # print(mask.dtype)
    # if i == 0:
    #   break
    yield (image,mask)
```

```
#Prepare-Validation
def augValidate():
  imageV_datagen = ImageDataGenerator()
  imageV_generator = imageV_datagen.flow_from_directory(validate_path,
                                                      classes = ['images'],
                                                      class_mode = None,
                                                      color_mode = 'rgb',
                                                      target_size = target_size,
                                                      batch_size = batch_size,
                                                      seed = seed,
                                                      )
  maskV_generator = imageV_datagen.flow_from_directory(validate_path,
                                                    classes = ['onlylabels'],
                                                    class_mode = None,
                                                    color_mode = 'grayscale',
                                                    target_size = target_size,
                                                    batch_size = batch_size,
                                                    seed = seed,
                                                    )
  validate_generator = zip(imageV_generator, maskV_generator)
  for (image,mask) in validate_generator:
    image = image / 255
    mask[mask > 0 ] = 1
    mask[mask == 0] = 0
    mask = np.squeeze(mask,axis = 3).astype(np.int32)
    yield (image,mask)
```

```
#model
filter_size = 3
pool_size = (2,2)
activation_function = ELU()
inputs = Input((240,320,3))

#loss&metric
def dice_coef(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + 1.) / (K.sum(y_true_f) + K.sum(y_pred_f) + 1.)


def dice_coef_loss(y_true, y_pred):
    return 1-dice_coef(y_true, y_pred)

conv1 = Conv2D(64, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(inputs)
 #BatchNormalization()
conv1 = Conv2D(64, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv1)
 #BatchNormalization()
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

conv2 = Conv2D(128, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(pool1)
 #BatchNormalization()
conv2 = Conv2D(128, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv2)
 #BatchNormalization()
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

conv3 = Conv2D(256, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(pool2)
 #BatchNormalization()
conv3 = Conv2D(256, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv3)
 #BatchNormalization()
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

conv4 = Conv2D(512, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(pool3)
 #BatchNormalization()
conv4 = Conv2D(512, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv4)
 #BatchNormalization()
drop4 = Dropout(0.5)(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2))(drop4)

conv5 = Conv2D(1024, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(pool4)
 #BatchNormalization()
conv5 = Conv2D(1024, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv5)
 #BatchNormalization()
drop5 = Dropout(0.5)(conv5)

upsample1 = UpSampling2D(size = (2,2))(drop5)
up6 = Conv2D(512, 2, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(upsample1)
merge6 = concatenate([drop4,up6], axis = 3)
conv6 = Conv2D(512, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(merge6)
 #BatchNormalization()
conv6 = Conv2D(512, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv6)
 #BatchNormalization()

upsample2 = UpSampling2D(size = (2,2))(conv6)
up7 = Conv2D(256, 2, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(upsample2)
merge7 = concatenate([conv3,up7], axis = 3)
conv7 = Conv2D(256, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(merge7)
 #BatchNormalization()
conv7 = Conv2D(256, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv7)
 #BatchNormalization()

upsample3 = UpSampling2D(size = (2,2))(conv7)
up8 = Conv2D(128, 2, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(upsample3)
merge8 = concatenate([conv2,up8], axis = 3)
conv8 = Conv2D(128, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(merge8)
 #BatchNormalization()
conv8 = Conv2D(128, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv8)
 #BatchNormalization()

upsample4 = UpSampling2D(size = (2,2))(conv8)
up9 = Conv2D(64, 2, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(upsample4)
merge9 = concatenate([conv1,up9], axis = 3)
conv9 = Conv2D(64, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(merge9)
 #BatchNormalization()
conv9 = Conv2D(64, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv9)
 #BatchNormalization()
conv9 = Conv2D(2, filter_size, activation = activation_function, padding = 'same', kernel_initializer = 'he_normal')(conv9)

conv10 = Conv2D(1, 1, activation = 'sigmoid')(conv9)

iou =keras.metrics.MeanIoU(num_classes=2)
model = Model(inputs = inputs, outputs = conv10)
model.compile(optimizer = Adam(lr = 0.001), loss = 'binary_crossentropy', metrics= 'accuracy')
model.summary()
```

```
train_set = augTrain()
validate_set = augValidate()
# model.fit(train_generator,validation_data=validate_generator,steps_per_epoch=120,validation_steps=25,epochs=10)
model.fit(train_set,validation_data=validate_set,steps_per_epoch=120,validation_steps=25,epochs=5)
```
